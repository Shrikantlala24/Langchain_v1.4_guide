```python
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
docs = [
    'deep learning is a subset of machine learning',
    'machine learning states mimicing human intelligence',
    'cat is Hungry of food',
    'dog is happy',
    'cats are very intelligent and mimic human behavior',
    'documentations are ridiculously important'
]
```

```python
load_dotenv()
model = HuggingFaceEmbeddings(
    model_name='sentence-transformers/all-MiniLM-L6-v2'
)

doc_emb = model.embed_documents(docs)
```

```python
query = 'Human'

query_emb = model.embed_query(query)
```

```python
print(cosine_similarity([query_emb], doc_emb)[0])
```

```python
def scoring(query, docs, doc_emb):
    query_emb = model.embed_query(query)
    
    similarity_scores = cosine_similarity([query_emb], doc_emb)[0]
    index, score=  sorted(list(enumerate(similarity_scores)),key=lambda x:x[1])[-1]

    print(docs[index])
    print('similarity score: ', score)
```

```python
scoring('documentations are important', docs, doc_emb)
```
